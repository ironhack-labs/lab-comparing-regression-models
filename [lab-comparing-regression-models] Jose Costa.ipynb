```python
#1. In this final lab, we will model our data. Import sklearn train_test_split and separate the data.
# importing X and Y from the previous lab

import pandas as pd

X=pd.read_csv("X.csv")
y=pd.read_csv("Y.csv")
X
```

```python
X.drop('Unnamed: 0', axis=1, inplace=True)
y.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4,random_state=100)
```

```python
#2. Try a simple linear regression with all the data to see whether we are getting good results.

lm=linear_model.LinearRegression()
model=lm.fit(X_train,y_train)
predictions=lm.predict(X_test)
r2_score(y_test,predictions)
```

This is a good r2 score actually!

```python
#3. Great! Now define a function that takes a list of models and train (and tests) them 
# so we can try a lot of them without repeating code.
# 4. Use the function to check LinearRegressor and KNeighborsRegressor.
#5. You can check also the MLPRegressor for this task!

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score
import warnings

def train_and_test_models(X_train, X_test, y_train, y_test, models, n_neighbors=5):
    # ignore warnings
    warnings.filterwarnings("ignore")
    
    for model in models:
        if type(model) == KNeighborsRegressor:
            model = KNeighborsRegressor(n_neighbors=n_neighbors)
        
        # Train the model on the training set
        model.fit(X_train, y_train)
        
        # Test the model on the testing set and calculate the mean squared error
        y_pred = model.predict(X_test)
        mse = mean_squared_error(y_test, y_pred)
        
        # Calculate the R-squared score
        r2 = r2_score(y_test, y_pred)
        
        # Print the model name, mean squared error and R-squared score
        print(f"{type(model).__name__}(n_neighbors={n_neighbors}): MSE={mse:.2f}, R2={r2:.2f}")

models = [LinearRegression(), KNeighborsRegressor(), MLPRegressor()]
train_and_test_models(X_train, X_test, y_train, y_test, models, n_neighbors=10)
```

```python
train_and_test_models(X_train, X_test, y_train, y_test, models, n_neighbors=50)
```

```python
train_and_test_models(X_train, X_test, y_train, y_test, models, n_neighbors=2000)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=50)
train_and_test_models(X_train, X_test, y_train, y_test, models, n_neighbors=2000)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=100)
train_and_test_models(X_train, X_test, y_train, y_test, models, n_neighbors=2000)
```
