# 1. In this final lab, we will model our data. Import sklearn train_test_split and separate the data.

```python
categorical_wrg = pd.read_csv('categorcial_wrg.csv',index_col=0)
numericals = pd.read_csv('numericals.csv')
y = pd.read_csv('y.csv',index_col=0)
```

```python
cat_encode = pd.DataFrame(OneHotEncoder().fit(pd.DataFrame(categorical_wrg)).transform(pd.DataFrame(categorical_wrg)).toarray())
```

```python
X = pd.DataFrame(np.concatenate((numericals, cat_encode), axis=1))
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4, random_state=100) 
```

# 2.Try a simple linear regression with all the data to see whether we are getting good results.

```python
model = sm.OLS(y_train, X_train).fit()
predictions = model.predict(X_test)
r2_score(y_test, predictions)
```

# 3. Great! Now define a function that takes a list of models and train (and tests) them so we can try a lot of them without repeating code.

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor

def model_evaluation(models_list, X_train, y_train, X_test, y_test):
    
    results = []
    
    for model in models_list:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)
        
        results.append({
            'model': type(model).__name__,
            'mse': mse,
            'r2': r2,
            'mae': mae
        })
        
    return results
```

```python
models = [LinearRegression(), KNeighborsRegressor(), MLPRegressor()]
results = model_evaluation(models, X_train, y_train, X_test, y_test)
```
