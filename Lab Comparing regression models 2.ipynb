```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("we_fn_use_c_marketing_customer_value_analysis.csv")
```

```python

missing_values = df.isnull().sum()


df_cleaned = df.dropna()
```

```python
df_cleaned
```

```python

df_cleaned = df_cleaned.drop(columns=['Customer'])
```

```python
numerical_features = df_cleaned.select_dtypes(include=['number'])
categorical_features = df_cleaned.select_dtypes(include=['object'])
```

```python
numerical_features.head()
```

```python
categorical_features.head()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error



y = df_cleaned['Customer Lifetime Value']


linear_reg_model = LinearRegression()
linear_reg_model.fit(numerical_features, y)
y_pred_linear = linear_reg_model.predict(numerical_features)
mse_linear = mean_squared_error(y, y_pred_linear)


def train_and_evaluate(models, X_train, y_train, X_test, y_test):
    results = {}
    predictions = {}
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        mse = mean_squared_error(y_test, y_pred)
        results[name] = mse
        predictions[name] = y_pred  # Store predictions for this model
    return results, predictions

# Define models
models = {
    'Linear Regression': LinearRegression(),
    'K-Nearest Neighbors': KNeighborsRegressor(),
    'MLP Regressor': MLPRegressor(random_state=42)
}


results, predictions = train_and_evaluate(models, numerical_features, y, numerical_features, y)


predictions_df = pd.DataFrame(predictions)


for name, mse in results.items():
    print(f'{name} Mean Squared Error: {mse}')


print(predictions_df)
```
